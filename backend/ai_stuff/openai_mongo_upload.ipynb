```python
import openai
import json
from pymongo import MongoClient
```

```python
with open('articles_metadata.json', 'r') as file:
    articles = json.load(file)
with open("secrets.json", 'r') as f:
    secrets = json.load(f)
```

```python
openai.api_key = secrets["openai_key"]
```

```python
# Preprocessing
to_embed = [json.dumps({
    "title": article["title"],
    "description": article["description"],
    "doc_type": article["doc_type"]
    })for article in articles]
```

```python
response = openai.embeddings.create(
    model="text-embedding-3-small",
    input=to_embed
)
```

```python
embedding_vectors = [point.embedding for point in response.data]
```

```python
len(embedding_vectors)
```

```python
secrets["mongo_uri"]
```

```python
import certifi
uri = secrets["mongo_uri"]

client = MongoClient(uri, tlsCAFile=certifi.where())
db = client["nasa"]
collection = db["article_collection"]


```

```python
to_upload = [{
    "title": article["title"],
    "description": article["description"],
    "doc_type": article["doc_type"],
    "url": article["url"]
    } for article in articles]
to_upload
```

```python
for doc, vector in zip(to_upload, embedding_vectors):
    doc["vector"] = vector  # Add the vector to the corresponding dictionary
```

```python
collection.insert_many(to_upload)
```
